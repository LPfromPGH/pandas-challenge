```python
#Load Libs, Find files, Read files

import pandas as pd
from pathlib import Path
schdatload = Path("Resources\schools_complete.csv")
studatload = Path("Resources\students_complete.csv")

schdat = pd.read_csv(schdatload)
studat = pd.read_csv(studatload)

schdatcomp = pd.merge(studat, schdat, how="left", on=["school_name", "school_name"])
schdatcomp.head()
schdatcomp.info()
```

```python
#Get the school count
schcnt = schdatcomp.value_counts(["school_name"])
schcnt
```

```python
#Get the student count
stucnt = len(schdatcomp)
stucnt
```

```python
#Total budget
totbud = schdatcomp['budget'].unique()
#print(totbud)
sumbud = totbud.sum()
print(sumbud)
```

```python
#Average math score
avgmath = schdatcomp['math_score'].mean()
print(avgmath)
```

```python
#Average reading score
avgread = schdatcomp['reading_score'].mean()
print(avgread)
```

```python
#Math Passers
passmath = schdatcomp[(schdatcomp["math_score"] >= 70)].count()["student_name"]
print(passmath)
passmathper = passmath / float(stucnt) * 100
print(passmathper)
```

```python
#What you readin for?
passread = schdatcomp[(schdatcomp["reading_score"] >= 70)].count()["student_name"]
print(passread)
passreadper = passread / float(stucnt) * 100
print(passreadper)
```

```python
# Passin both
mathreadcnt = schdatcomp[(schdatcomp["math_score"] >= 70) & (schdatcomp["reading_score"] >= 70)].count()["student_name"]
print(mathreadcnt)
mathreadavg = mathreadcnt / float(stucnt) * 100
print(mathreadavg)
```

```python
#School count
totsch = schdatcomp["school_name"].unique()
print(totsch)
schtot = len(totsch)
print(schtot)
```

```python
#making the first data frame to output
firstout = pd.DataFrame([[schtot, stucnt, sumbud, avgmath, avgread, passmathper, passreadper, mathreadavg]], columns = ['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing'])
firstout["Total Students"] = firstout["Total Students"].map("{:,}".format)
firstout["Total Budget"] = firstout["Total Budget"].map("${:,.2f}".format)
firstout
```

```python
#Grouping by school
schgrp = schdatcomp.groupby(["school_name"])
```

```python
#school types
schtype = schgrp["type"].first()
schtype
```

```python
#Student count by school
grptotstu = schgrp["Student ID"].count()
grptotstu
```

```python
#budget by school
totschbud = schgrp["budget"].first()
totschbud
```

```python
#budget per studet
percapbud = totschbud / grptotstu
percapbud
```

```python
#avg math score by school
grpmathavg = schgrp["math_score"].mean()
grpmathavg
```

```python
#avg reading score by school
grpreadavg = schgrp['reading_score'].mean()
grpreadavg
```

```python
#count of math passers by school
grpmathpass = schdatcomp[schdatcomp["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
grpmathpass
```

```python
#count of read passers by school
grpreadpass = schdatcomp[schdatcomp["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
grpreadpass
```

```python
#Percentage math passers by school
perpassmath = grpmathpass / grptotstu * 100
perpassmath
```

```python
#Percent reading passers by school
perpassread = grpreadpass / grptotstu * 100
perpassread
```

```python
#overall passing by school
overallpass = (grpmathpass + grpreadpass) / 2
overallpassper = overallpass / grptotstu * 100
overallpassper
```

```python
#Building school summary data frame
schoolsumm = pd.DataFrame({"School Type": schtype, "Total Students": grptotstu, "School Budget": totschbud, "Per Capita Budget": percapbud, "Avg Math Score": grpmathavg, "Avg Reading Score": grpreadavg, "% Passing Math": perpassmath, "% Passing Reading": perpassread, "% Overall Pass": overallpassper})

schoolsumm["School Budget"] = schoolsumm["School Budget"].map("${:,.2f}".format)
schoolsumm["Per Capita Budget"] = schoolsumm["Per Capita Budget"].map("${:,.2f}".format)
schoolsumm
```

```python
#Sorting the cream of the crop
top_school = schoolsumm.sort_values(["% Overall Pass"], ascending=False)
top_school.head()
```

```python
#Sorting the bottom of the barrel
bottom_school = schoolsumm.sort_values(["% Overall Pass"], ascending=True)
bottom_school.head()
```

```python
#Avg math and reading scores by grade and by school
ninth_graders = schdatcomp[(schdatcomp["grade"] == "9th")]
tenth_graders = schdatcomp[(schdatcomp["grade"] == "10th")]
eleventh_graders = schdatcomp[(schdatcomp["grade"] == "11th")]
twelfth_graders = schdatcomp[(schdatcomp["grade"] == "12th")]

ninth_math = ninth_graders.groupby(["school_name"])["math_score"].mean()
tenth_math = tenth_graders.groupby(["school_name"])["math_score"].mean()
eleventh_math = eleventh_graders.groupby(["school_name"])["math_score"].mean()
twelfth_math = twelfth_graders.groupby(["school_name"])["math_score"].mean()
ninth_read = ninth_graders.groupby(["school_name"])["reading_score"].mean()
tenth_read = tenth_graders.groupby(["school_name"])["reading_score"].mean()
eleventh_read = eleventh_graders.groupby(["school_name"])["reading_score"].mean()
twelfth_read = twelfth_graders.groupby(["school_name"])["reading_score"].mean()
```

```python
#Building data frame for math scores by grade and school
math_scores_by_grade = pd.DataFrame({"9th Grade Math": ninth_math, "10th Grade Math": tenth_math, "11th Grade Math": eleventh_math, "12th Grade Math": twelfth_math})
math_scores_by_grade.head(15)
```

```python
#Building data frame for reading scores 
read_scores_by_grade = pd.DataFrame({"9th Grade Reading": ninth_read, "10th Grade Reading": tenth_read, "11th Grade Reading": eleventh_read, "12th Grade Reading": twelfth_read})
read_scores_by_grade.head(15)
```
